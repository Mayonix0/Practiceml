```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset=pd.read_csv('Train.csv')
X_train=np.ones((699,2))
X_train[:,1]=dataset['x']#making the x column of dataset the second column for X_train 
Y_train=dataset['y']#making the y column of dataset equal to Y_train
features=dataset.shape[1]#no of features 
training_examples=dataset.shape[0]#no of training examples
m=training_examples#legal term for training examples is m
theta=np.zeros(features)#an array for parameters w and b
n_iter=100 #no of iteration for which gradient descent algorithm will work
alpha=0.0001#learning rate
for i in range(n_iter):
        derivative=np.zeros(features)
        y_pred=np.dot(X_train,theta)
        error=y_pred-Y_train
        for j in range(features):
            #derivative for w will be error *x
            #derivative for b will be error *1
            derivative[j]=np.sum(np.dot(error,X_train)) 
        theta=theta-(1/m)*derivative*alpha#gradient descent algorithm 
finaly=theta[0]+theta[1]*X_train[:,1]#the y at which cost function is minimum is found and which will best fit 
#plot the output
plt.scatter(X_train[:,1],Y_train)
plt.plot(X_train[:,1],finaly,color="orange")
#end of code
        
```
